```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Covide19India.csv')
data
```

```python
Day=data[data['Current Status']=='Hospitalized'].groupby(['Month','Day'])['Num Cases'].sum()
Day
```

```python
y=Day.values
y
```

```python
x=np.arange(len(Day))
x
```

```python
x=x.reshape(-1,1)
y=y.reshape(-1,1)
x
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
sc_y=StandardScaler()
Sx=sc_x.fit_transform(x)
Sy=sc_y.fit_transform(y)
```

```python
Sx
```

```python
Sy
```

```python
from sklearn.svm import SVR
reg=SVR(kernel='rbf')
reg.fit(Sx,Sy.ravel())
```

```python
reg.score(Sx,Sy)*100
```

```python
plt.scatter(Sx,Sy)
plt.plot(Sx,reg.predict(Sx),color='k',linewidth=5)
plt.show()
```
