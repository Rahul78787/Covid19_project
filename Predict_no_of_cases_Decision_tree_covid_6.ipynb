```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Covide19India.csv')
data
```

```python
  Day=data[data['Current Status']=='Hospitalized'].groupby(['Month','Day'])['Num Cases'].sum()
Day
```

```python
y=Day.values
y
```

```python
x=np.arange(len(Day))
x=x.reshape(-1,1)
x
```

```python
from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
Yp=reg.predict(x)
Yp
```

```python
plt.scatter(x,y)
plt.plot(x,Yp,color='k',linewidth=2)
plt.show()
```

```python
reg.score(x,y)*100
```
