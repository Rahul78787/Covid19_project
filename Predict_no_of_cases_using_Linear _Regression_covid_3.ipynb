```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Covide19India.csv')
data
```

```python
data
```

```python
data=data.iloc[1:,:]
data
```

```python
data=data.iloc[:,1:]
data
```

```python
data[data['Current Status']=='Hospitalized'].groupby(['Month','Day'])['Num Cases'].sum()
```

```python
Day=data[data['Current Status']=='Hospitalized'].groupby(['Month','Day'])['Num Cases'].sum()
len(Day)
```

```python
x=np.arange(len(Day))
x
```

```python
Day
```

```python
y=Day.values
y
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x
```

```python
x=x.reshape(-1,1)
x
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x,y)
```

```python
y[308]
```

```python
regressor.predict([[309]])
```

```python
y
```

```python
yp=regressor.predict(x)
yp
```

```python
plt.scatter(x,y)
plt.plot(x,yp)
plt.show()
```

```python
plt.plot(x,yp)
```

```python
regressor.score(x,y)*100
```

```python
regressor.intercept_
```
